```python
import csv, os
import pandas as pd
```

```python
df = pd.read_csv('datasets\\good_recs.csv', skiprows=0, header=0)
```

```python
#df
#df['exam_1'].count()
#df.count()
```

```python
#1.1 Make a copy of the DataFrame and call it df_copy (we will use the copy for all exercises)
df_copy = df.copy()
```

```python
#1.2 Find all records where â€˜exam_2â€™ <= 60.0
df_copy[df_copy['exam_2'].le(60.0)]
```

```python
#1.3 List the indexes for the records in question #1.2
df_copy[df_copy['exam_2'].le(60.00)].index
```

```python
#1.4 Find all records where â€˜exam_1â€™ and â€˜exam_2â€™ are >= 65.0
#apply boolean mask to dataframe
filter = (df_copy['exam_1']>=65.0) & (df_copy['exam_2']>=65.0)
df_copy[filter]
#df_copy[(df_copy.exam_1 >=65.0) & (df_copy.exam_2 >=65.0)]
```

```python
#1.5 List the indexes for the records in q. 1.4
df_copy[filter].index
```

```python
#1.6 Find all the records where â€˜exam_1â€™ is NaN (null)
df_copy[df_copy['exam_1'].isnull()]
```

```python
#1.7 List the indexes for records in q. 1.6
df_copy[df_copy['exam_1'].isnull()].index
```

```python
#1.8 Find all records where both â€˜exam_1â€™ and â€˜exam_2â€™ are null. 
filter = df_copy['exam_1'].isnull() & df_copy['exam_2'].isnull()
df_copy[filter]
```

```python
#1.9 Drop records where â€˜exam_3â€™ is null
df_copy.dropna(subset = ['exam_3'],inplace=False)
```

```python
#1.10 Change the value of NaN (nulls) to 49.0
df_copy.fillna(value = 49.0, inplace=False).head(11)
```

```python
#1.11 Drop the column â€˜transcodeâ€™
df_copy.drop('transcode', axis=1, inplace=False)
```

```python
#1.12 Change the value of exam_1 to 100.0 for record 3 
df_copy.loc[2,'exam_1'] = 100.00
df_copy.head(5)
```

```python
df_copy.head(10)
```

```python
#1.13 Substract 2 from every exam_2 value
df_copy['exam_2'] = df_copy['exam_2'] - 2
df_copy.head(10)
```

```python
#1.14 Add a new column â€˜AVGâ€™ with mean of all 3 exams.
df_copy['AVG'] = df_copy[['exam_1', 'exam_2', 'exam_3']].mean(axis=1).round(2)
```

```python
df_copy.head(10)
```

```python
#1.15 What is the Max and Min values for exam_2

print('Minimum value for exam_2:' , df['exam_2'].min())
print('Maximum value for exam_2:' , df['exam_2'].max())
```

```python
#1.16 What is the Standard Deviation for exam_1, exam_2 and exam_3 (separately)

print('Standard deviation for exam_1:' , "{:.2f}".format(df['exam_1'].std()))
print('Standard deviation for exam_2:' , "{:.2f}".format(df['exam_2'].std()))
print('Standard deviation for exam_3:' , "{:.2f}".format(df['exam_3'].std()))
```

```python
#1.17 Group records by â€˜Deptâ€™ and show the mean for exam_1, exam_2 and exam_3
dept_group = df_copy.groupby('Dept')
dept_group[['exam_1','exam_2', 'exam_3']].mean().round(2)
```
